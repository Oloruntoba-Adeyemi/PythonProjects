```python
#Q14
def replace_outliers_with_mean(column):
    mean = column.mean()
    std = column.std()
    lower_bound = mean - 3 * std
    upper_bound = mean + 3 * std
    column = column.apply(lambda x: x if lower_bound <= x <= upper_bound else mean)
    return column

df['marks'] = replace_outliers_with_mean(df['marks'])
df['age'] = replace_outliers_with_mean(df['age'])

df.describe()
```

```python
#25 Add another column 'Grades'. If marks are greater than 70, set grade as A, if its between 60 and 70,
#set grade as B, if its between 50 and 60, grade as C, otherwise set Grade as F



def grade(mark):
    if mark > 70:
        return 'A'
    elif 60 <= mark <= 70:
        return 'B'
    elif 50 <= mark < 60:
        return 'C'
    else:
        return 'F'


df['grades'] = [grade(mark) for mark in df['marks']]

```
